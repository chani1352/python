```python
# Lotto 복권 코딩 11
## 1 ~ 45 숫자를 중복없이 6개 뽑고, 보너스 번호를 중복없이 1개를 뽑는다
### 1등: 6개 숫자, 2등: 5개 숫자 + 보너스, 3등: 5개 숫자, 4등: 4개, 5등: 3개 
import random

# 복권 시스템 클래스
class LottoSystem:
    def __init__(self, num_tickets=10):
        self.num_tickets = num_tickets
        self.tickets = []
        self.winning_numbers = set()
        self.results = {
            "1등": [],
            "2등": [],
            "3등": [],
            "4등": [],
            "5등": []
        }
    #복권 발행
    def generate_lotto_tickets(self):
        ticket = set()
        for _ in range(self.num_tickets):
          a = True
          while a :
            if len(ticket) < 6 :
              ticket.add((random.randint(1,45)))
            else : 
              a = False
          self.tickets.append(ticket)
          ticket = set()
        return self.tickets
        # for _ in range(self.num_tickets):
        #   self.tickets.append(sorted(random.sample(range(1,46),6)))
        # return self.tickets

    # 당첨 복권
    def draw_winning_numbers(self):
      a = True
      while a :
        if len(self.winning_numbers) < 7 :
          self.winning_numbers.add((random.randint(1,45)))
        else : 
          a = False
      return self.winning_numbers
      # winning_numbers = random.sample(range(1,46),7)
      # bonus_number = winning_numbers[-1:]
      # self.winning_numbers = sorted(winning_numbers[:6]) + bonus_number
      # print(self.winning_numbers)
      # return self.winning_numbers

    # 당첨 처리 
    def check_winners(self):
        for ticket in self.tickets:
            self.check_ticket(ticket)
        self.display_ticket_with_matches()
                   
    # 당첨 분류
    def check_ticket(self, ticket):
        bonus_matched = False
        result = ticket & self.winning_numbers
        if len(result) == 6 :
            if self.winning_numbers[-1] in ticket:
                bonus_matched = True
        if len(result) == 6:
            self.results["1등"].append(ticket)
        elif len(result) == 6 and bonus_matched:
            self.results["2등"].append(ticket)
        elif len(result) == 5:
            self.results["3등"].append(ticket)
        elif len(result) == 4:
            self.results["4등"].append(ticket)
        elif len(result) == 3:
            self.results["5등"].append(ticket)

    # # 당첨복권 출력
    def display_ticket_with_matches(self):
        ranks = ["1등","2등","3등","4등","5등"]
        display_ticket = []
        print(f"당첨 번호 (6개 번호 + 보너스 번호) :{self.winning_numbers}")
        print(f"티켓 구매 수 : {self.num_tickets}장")
        for rank in ranks :
            if len(self.results[rank]) > 0 :
                print(f'{rank} 당첨 티켓 : {len(self.results[rank])}장 (확률 : {round(len(self.results[rank])/self.num_tickets*100,2)}%)')
                for ticket in self.results[rank] :
                    print(f"복권 번호 : ",end="")
                    for num in ticket :
                        if num in self.winning_numbers[:6]:
                            print(f"[{num}]",end=" ")
                        else : 
                            print(num,end=" ")
                    print()

# 실행 함수

lotto_system = LottoSystem(num_tickets=10000)
lotto_system.generate_lotto_tickets()
lotto_system.draw_winning_numbers()
lotto_system.check_winners()
```
