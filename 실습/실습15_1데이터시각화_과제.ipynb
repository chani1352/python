```python
import numpy as np
import matplotlib.pyplot as plt
############### 
### plt.title('군집 산포도') 출력을 위한 한글 폰트 설정
# 한글 폰트 경로 설정 (예: 맑은 고딕)
#font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows의 경우
# 폰트 설정
#font_prop = fm.FontProperties(fname=font_path, size=12)
#plt.rc('font', family=font_prop.get_name())
#############

# 임의의 군집(클러스터) 데이터를 생성하는 함수 => 이 함수를 이해하는 것이 중요
def generate_cluster(center, num_points, spread):
    # 정규 분포를 따르는 점들을 생성
    points = np.random.randn(num_points, 2) * spread + center # 난수를 num_points 갯수 만큼, 2차원으로 생성
    return points
# np.random.randn(d0, d1, ..., dn)은 주어진 차원(d0, d1, ..., dn)의 배열을 생성
# num_points: 배열의 첫 번째 차원의 크기, 즉 생성할 데이터 포인트의 개수
# 2: 배열의 두 번째 차원의 크기, 즉 각 데이터 포인트의 차원  즉 x,y 2차원   randn(400,2) 400은 행의 수 2는 열의 수

# 각 군집의 중심, 점 개수, 확산 정도 설정
cluster_1 = generate_cluster(center=[2, 3], num_points=100, spread=0.9) # center 중심점 num~ 점의 개수 spread 난수 생성시 퍼지는 정도 0.1이면 많이 모여있음 0.9면 퍼져있음
cluster_2 = generate_cluster(center=[7, 8], num_points=200, spread=0.8)
cluster_3 = generate_cluster(center=[5, 2], num_points=300, spread=0.3)

print(cluster_1[:5])
print(type(cluster_1))

# 각 군집을 하나의 배열로 합침
all_points = np.vstack((cluster_1, cluster_2, cluster_3))
# NumPy의 vstack 함수를 사용하여 여러 배열을 수직으로 쌓아 새로운 배열을 생성하는 함수
# 군집을 시각화
plt.figure(figsize=(8, 6))

# 첫 번째 군집을 파란색으로
plt.scatter(cluster_1[:, 0], cluster_1[:, 1], color='blue', label='Cluster 1', alpha=0.6)

# 두 번째 군집을 빨간색으로
plt.scatter(cluster_2[:, 0], cluster_2[:, 1], color='red', label='Cluster 2', alpha=0.6)

# 세 번째 군집을 녹색으로
plt.scatter(cluster_3[:, 0], cluster_3[:, 1], color='green', label='Cluster 3', alpha=0.6)

# 그래프 설정
plt.title('군집 산포도')
plt.xlabel('X-축')
plt.ylabel('Y-축')
plt.legend()
plt.grid(True)

# 그래프 출력
plt.show()
```

```python
from random import choice


class RandomWalk:
    """A class to generate random walks."""

    def __init__(self, num_points=5000):
        """Initialize attributes of a walk."""
        self.num_points = num_points

        # All walks start at (0, 0).
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """Calculate all the points in the walk."""
        # Keep taking steps until the walk reaches the desired length.
        while len(self.x_values) < self.num_points:
            ###################
            ### 동서남북 대각선 8가지 방향으로 이동하는 코드 작성
            # 이동 방향을 정의
            directions = [
                (1, 0),    # 동쪽   0
                (-1, 0),   # 서쪽   1  
                (0, 1),    # 북쪽   2
                (0, -1),   # 남쪽   3
                (1, 1),    # 북동쪽 4
                (-1, 1),   # 북서쪽 5
                (1, -1),   # 남동쪽 6
                (-1, -1)   # 남서쪽 7
            ]

            #8가지 방향을 random하게 선택
         
            ### 코딩할 부분
            ###############################

            d = choice([0,7])
            x_distance = choice([0, 1, 2, 3, 4])
            y_distance = choice([0, 1, 2, 3, 4])
            x_step = x_distance * directions[d][0]
            y_step = y_distance * directions[d][1]

            # Decide which direction to go, and how far to go.
            # x_direction = choice([1, -1])   #1 -1 중에 랜덤으로 choice
            # x_distance = choice([0, 1, 2, 3, 4])
            # x_step = x_direction * x_distance

            # y_direction = choice([1, -1])
            # y_distance = choice([0, 1, 2, 3, 4])
            # y_step = y_direction * y_distance

            # Reject moves that go nowhere.
            if x_step == 0 and y_step == 0:
                continue

            # Calculate the new position.
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step

            self.x_values.append(x)
            self.y_values.append(y)
```

```python
import matplotlib.pyplot as plt

#from random_walk import RandomWalk

# Keep making new walks, as long as the program is active.
while True:
    # Make a random walk.
    rw = RandomWalk(10)
    rw.fill_walk()

    # Plot the points in the walk.
    plt.style.use('classic')
    fig, ax = plt.subplots()
    point_numbers = range(rw.num_points)
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues,
       edgecolors='none', s=10)
    ax.set_aspect('equal')

    # Emphasize the first and last points.
    ax.scatter(0, 0, c='green', edgecolors='none', s=100)
    ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none',
        s=100)

    # Remove the axes.
    ax.get_xaxis().set_visible(True)
    ax.get_yaxis().set_visible(True)
    
    plt.show()

    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

```python
### 벌레 이동 코딩 문제 구현 >> 
### 체스에 0이 없어질 때까지 이동 >> 바둑판 19 x19 으로 수정하는 코딩 
import numpy as np
import random
import matplotlib.pyplot as plt

# 10x10 배열 생성 (초기 값은 모두 0)
grid = np.zeros((8, 8), dtype=int)

# 벌레의 초기 위치 (임의로 설정)
x, y = random.randint(0, 7), random.randint(0, 7)

# 벌레의 이동 방향 (동서남북 및 대각선 8방향)
directions = [
    (-1, 0), (1, 0),  # 남, 북
    (0, -1), (0, 1),  # 서, 동
    (-1, -1), (-1, 1),  # 남서, 남동
    (1, -1), (1, 1)  # 북서, 북동
]

# 반복 실행
steps = 0
while np.any(grid == 0): #grid 배열에 0이 하나라도 있으면 true
    # 현재 위치에 1 추가
#### 구현할 코드
    d = choice(range(0,8))
    if  0 <= (x+directions[d][1]) <= 7 and 0 <= (y+directions[d][0]) <= 7 :
        x += directions[d][1]
        y += directions[d][0]        
        grid[x][y] +=1
        steps += 1


# 결과 출력
print(f"총 {steps} 번 이동 후 배열이 채워졌습니다.")

# 배열의 결과를 시각화 (그래프 출력)
plt.imshow(grid, cmap='viridis', interpolation='none')
plt.colorbar(label="방문횟수")
plt.title("벌레의 이동 경로: 바둑판  10x10 Grid")
plt.show()
```

```python
from random import randint
# Die 클래스를 정의 (6면체 주사위)
class Die:
    def __init__(self):
        self.num_sides = 6  # 주사위 면의 수

    # 주사위를 던지는 함수 (1-6 사이 임의 값을 반환)
    def roll(self):
        return randint(1, self.num_sides)
```

```python
import plotly.express as px

# Create a D6.
die = Die() #Dice는 주사위 복수개 

# Make some rolls, and store results in a list.
results = []
for roll_num in range(1000):
    result = die.roll()
    results.append(result)

# Analyze the results.
frequencies = []
poss_results = range(1, die.num_sides+1)
for value in poss_results:
    frequency = results.count(value)
    frequencies.append(frequency)

# Visualize the results.
title = "주사위 1,000회 결과"
labels = {'x': '주사위값', 'y': '빈도수'}
fig = px.bar(x=poss_results, y=frequencies, title=title, labels=labels)
fig.show()
```

```python
import random
import matplotlib.pyplot as plt
#######################
# 한글 폰트 경로 설정 (예: 맑은 고딕)
#font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows의 경우

# 폰트 설정
#font_prop = fm.FontProperties(fname=font_path, size=12)
#plt.rc('font', family=font_prop.get_name())
#######################

# 주사위 3개를 100번 던져서 합을 구하는 함수
def roll_dice(num_trials):
    die1 = Die()
    die2 = Die()
    die3 = Die()
    
    results = []
    for _ in range(num_trials):
        total = die1.roll() + die2.roll() + die3.roll()
        results.append(total)
    
    return results

# 100번 시행
num_trials = 100
results = roll_dice(num_trials)

# 결과의 빈도 계산
poss_results = range(3, 19)  # 가능한 주사위의 합 값 (3~18)
frequencies = [results.count(x) for x in poss_results]  # 각 합의 빈도 계산

# 막대 그래프 시각화
plt.bar(poss_results, frequencies, color='blue', alpha=0.7)
plt.title(f"3개 주사위 값의 합을 {num_trials}번 반복한 막대차트")
plt.xlabel("3개 주사위 값의 합")
plt.ylabel("빈도수")
plt.xticks(poss_results)  # x축의 값 설정

# 그래프 보여주기
plt.grid(axis='y')
plt.show()
```

```python
# Lotto 복권 코딩 11
## 1 ~ 45 숫자를 중복없이 6개 뽑고, 보너스 번호를 중복없이 1개를 뽑는다
### 1등: 6개 숫자, 2등: 5개 숫자 + 보너스, 3등: 5개 숫자, 4등: 4개, 5등: 3개 
import random

# 복권 시스템 클래스
class LottoSystem:
    def __init__(self, num_tickets=10):
        self.num_tickets = num_tickets
        self.tickets = []
        self.winning_numbers = []
        self.results = {
            "1등": [],
            "2등": [],
            "3등": [],
            "4등": [],
            "5등": []
        }
    #복권 발행
    def generate_lotto_tickets(self):
        for _ in range(self.num_tickets):
            self.tickets.append(sorted(random.sample(range(1,46),6)))
        return self.tickets

    # 당첨 복권
    def draw_winning_numbers(self):
        winning_numbers = random.sample(range(1,46),7)
        bonus_number = winning_numbers[-1:]
        self.winning_numbers = sorted(winning_numbers[:6]) + bonus_number
        print(self.winning_numbers)
        return self.winning_numbers

    # 당첨 처리 
    def check_winners(self):
        for ticket in self.tickets:
            self.check_ticket(ticket)
        self.display_ticket_with_matches()
                   
    # 당첨 분류
    def check_ticket(self, ticket):
        matched_count, bonus_matched = 0,False
        for i in ticket :
            if i in self.winning_numbers[:6]:
                matched_count +=1
        if matched_count ==5 :
            if self.winning_numbers[-1] in ticket:
                bonus_matched = True
        if matched_count == 6:
            self.results["1등"].append(ticket)
        elif matched_count == 5 and bonus_matched:
            self.results["2등"].append(ticket)
        elif matched_count == 5:
            self.results["3등"].append(ticket)
        elif matched_count == 4:
            self.results["4등"].append(ticket)
        elif matched_count == 3:
            self.results["5등"].append(ticket)

    # # 당첨복권 출력
    def display_ticket_with_matches(self):
        ranks = ["1등","2등","3등","4등","5등"]
        display_ticket = []
        print(f"당첨 번호 (6개 번호 + 보너스 번호) :{self.winning_numbers}")
        print(f"티켓 구매 수 : {self.num_tickets}장")
        for rank in ranks :
            if len(self.results[rank]) > 0 :
                print(f'{rank} 당첨 티켓 : {len(self.results[rank])}장 (확률 : {len(self.results[rank])/self.num_tickets*100}%)')
                for ticket in self.results[rank] :
                    print(f"복권 번호 : ",end="")
                    for num in ticket :
                        if num in self.winning_numbers[:6]:
                            print(f"[{num}]",end=" ")
                        else : 
                            print(num,end=" ")
                    print()

# 실행 함수

lotto_system = LottoSystem(num_tickets=50000)
lotto_system.generate_lotto_tickets()
lotto_system.draw_winning_numbers()
lotto_system.check_winners()
```
